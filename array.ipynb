```python
# 1. Two Sum

# given an array of integer snums and an integer target,
# return indices of the two numbers such that they add up to target.

# you may assume that each input would have exactly one solution, and you may not use the same element twice. 
# you can return the answer in any order

class Solution:
  def twoSum(self, nums: List[int], target: int) -> List[int]:
    hash = dict()
    
    for index, number in enumerate(nums):
      if number not in hash:
        hash[number] = index
      if target - number in hash and hash[target - number] != index:
        return [index, hash[target - number]]

# time complexity O(n), Space complexity O(n)

# runtime: 68ms, beats 82%
# memory: 15MB, beats 66%
```

```python
# 350. Intersection of Two Arrays II
# Given two integer arrays nums1 and nums2, return an array of their intersection. 
# Each element in the result must appear as many times as it shows in both arrays and you may return the result in any order.

class Solution:
    def intersect(self, nums1: List[int], nums2: List[int]) -> List[int]:
        hash = dict()
        for num in nums1:
            if num not in hash:
                hash[num] = [0,0]
            hash[num][0] += 1
        
        for num in nums2:
            if num in hash:
                hash[num][1] += 1
        
        output = []

        for key, array in hash.items():
            for n in range(min(array)):
                output.append(key)
        return output
  
  # time O(n), space O(n + m)
  
  # runtime: 39ms, beats 99%
  # memory: 14MB, beats 51%
```

```python
# 121. Best Time to Buy and Sell Stock
# You are given an array prices where prices[i] is the price of a given stock on the ith day.

# You want to maximize your profit by choosing a single day to buy one stock and choosing a different day in the future to sell that stock.

# Return the maximum profit you can achieve from this transaction. If you cannot achieve any profit, return 0.


class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        min_stock_price = prices[0]
        max_profit = 0

        for stock in prices[1:]:
            if min_stock_price > stock:
                min_stock_price = stock
            elif min_stock_price < stock:
                if max_profit < stock - min_stock_price:
                    max_profit = stock - min_stock_price
        return max_profit 


# time O(n), space O(1)
# runtime: 963ms, beats 97%
# memory: 25MB, beats 38%
```

```python
# 118. Pascal's Triangle
# Given an integer numRows, return the first numRows of Pascal's triangle.

# In Pascal's triangle, each number is the sum of the two numbers directly above it as shown:

# Input: numRows = 5
# Output: [[1],[1,1],[1,2,1],[1,3,3,1],[1,4,6,4,1]]

# Input: numRows = 1
# Output: [[1]]

class Solution:
    def generate(self, numRows: int) -> List[List[int]]:
      output = []
      for i in range(numRows):
        subarray = []
        for j in range(i + 1):
          if j == 0 or i + 1 == j + 1:
            subarray.append(1)
          else:
            subarray.append(output[i-1][j] + output[i-1][j-1])
        output.append(subarray)
      return output
          
            
  # time: O(n^2) number of rows ^2, space: O(1), the input and output do not generally count towards space complexity
  # runtime: 47ms, beats 57%
  # memory: 13MB, beats 67%
```

```python
# 74. Search a 2D Matrix

# You are given an m x n integer matrix matrix with the following two properties:

# Each row is sorted in non-decreasing order.
# The first integer of each row is greater than the last integer of the previous row.
# Given an integer target, return true if target is in matrix or false otherwise.

# You must write a solution in O(log(m * n)) time complexity.


class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:

        left = 0
        right = len(matrix)
        while left < right:
            middle = left + (right - left) // 2
            middle_arr = matrix[middle]
            if middle_arr[0] <= target <= middle_arr[-1]:
                if target in set(middle_arr):
                    return True
                else:
                    return False
 
            elif middle_arr[0] > target:
                right = middle 
            elif middle_arr[-1] < target:
                left = middle + 1
  # time O(log(n)), space O(1)
```

```python
# 387. First Unique Character in a String
# Given a string s, find the first non-repeating character in it and return its index. If it does not exist, return -1.

class Solution:
    def firstUniqChar(self, s: str) -> int:
        cache = dict()
        for index in range(len(s)):
            if s[index] not in cache:
                cache[s[index]] = [index, 1]
            else:
                cache[s[index]][1] += 1
        for key, arr in cache.items():
            if arr[1] == 1:
                return arr[0]
        return -1

# O(n) time, space O(n)
```

```python
# 383. Ransom Note
# Given two strings ransomNote and magazine, return true if ransomNote can be constructed by using the letters from magazine and false otherwise.

# Each letter in magazine can only be used once in ransomNote.

class Solution:
    def canConstruct(self, ransomNote: str, magazine: str) -> bool:
        cache = dict()
        for char in ransomNote:
            if char not in cache:
                cache[char] = 0
            cache[char] += 1
        
        for char in magazine:
            if char in cache:
                cache[char] -=1
                if cache[char] == 0:
                    del cache[char]
                if cache == {}:
                    return True
        return False

# time O(n), space(n)
```
